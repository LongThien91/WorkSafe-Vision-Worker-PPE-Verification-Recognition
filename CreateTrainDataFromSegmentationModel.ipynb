Import Model and Segment Data

```python
modelSeg = YOLO('/content/gdrive/MyDrive/segmentlbest.pt')
```

```python
filenames = [f for f in os.listdir('/content/DatasetVestAndHelmet/train/images') if f.endswith(('.jpg', '.jpeg', '.png'))]
```

```python
#create data
save_folder_path = '/content/DataForTraining'
counter=1
```

```python
for file in filenames:
  image_path='/content/DatasetVestAndHelmet/train/images/'+file
  image=cv2.imread(image_path)
  image=cv2.resize(image,(512,512))

  results=modelSeg.predict(source=image,conf=0.3, save=False)
  for i in range(0,len(results[0])):
    mask=results[0].masks.data[i].numpy()
    mask_image = image.copy()
    mask_image[mask == 0] = [255, 255, 255]
    img_name = f'Human{counter}'
    new_filename = f"{img_name}.jpg"
    counter=counter+1

    new_image_path = os.path.join(save_folder_path, new_filename)
    cv2.imwrite(new_image_path, mask_image)
```
